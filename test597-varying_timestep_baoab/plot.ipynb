```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
samples = np.fromfile("samples.dat", sep=" ").reshape(-1, 2)
```

```python
n = len(samples)

fig = plt.figure(figsize=(9, 2), dpi=96, tight_layout=True)
grid = plt.GridSpec(1, 2, width_ratios=[8, 1])

ax = fig.add_subplot(grid[0])
ax.grid(True, ls=":", zorder=1)
ax.plot(samples[:n, 0], samples[:n, 1], lw=0.25, zorder=2)
ax.set_xlim(samples[0, 0], samples[n - 1, 0])
ax.set_ylim(-30, 30)
ax.set_xticklabels([f"{x:.0f}" for x in ax.get_xticks()], rotation=10, ha="right")

ax = fig.add_subplot(grid[1])
ax.grid(True, ls=":", zorder=1)
ax.hist(samples[:n, 1], bins=np.linspace(-30, 30, num=100), orientation="horizontal", zorder=2)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_ylim(-30, 30)

pass
```

```python
k = 0.01
epsilon = 0.5
wavenumber = 0.8
temperature = 1

bins = np.linspace(-35, 35, num=100)
x = np.linspace(-35, 35, num=1000)

u = k/2 * x**2 + epsilon * np.sin(wavenumber * x)

p = np.exp(-u / temperature)
p *= len(samples) / p.sum()

fig, ax = plt.subplots(figsize=(5, 3), dpi=96)

ax.grid(True, ls=":", zorder=1)
ax.hist(samples[:, 1], bins=x, zorder=2)
ax.plot(x, p, zorder=3)

pass
```
